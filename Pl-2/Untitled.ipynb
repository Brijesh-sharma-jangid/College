```python
import easyocr
```

```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
images = []
```

```python
for file in os.listdir('./Data/'):
    image_path = os.path.join('./Data/', file)
    img = cv2.imread(image_path)
    if img is not None:
        images.append(img)
```

```python
plt.imshow(images[0])
```

```python
reader = easyocr.Reader(['en'],gpu=False) 
```

```python
results = reader.readtext(images[0])
```
